### Word Cloud

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def generate_clean_text(text_list):
    
    all_text = ' '.join(line for line in text_list if line not in ['<Media omitted>'] and type(line)==str)
    eg_stopwords = set(stopwords.words('english'))
    garbage = ['s', 't', 'n\'t', '\'s', '\'ll']
    
    tokens = word_tokenize(all_text)
    cleaned_text = ' '.join(word.lower() for word in tokens if word not in eg_stopwords and word not in garbage)
    return cleaned_text
```

### Message frequency among weekdays and weekends

```python
import datetime

def is_weekend(date):
    # Get the day of the week (Monday is 0, Sunday is 6)
    day_of_week = date.weekday()
    
    # Check if it's Saturday (5) or Sunday (6)
    return day_of_week >= 5
```

### Conversation initiation analysis

#### Generate df

```python
import pandas as pd
import re
def generate_dataframe(file_path):
    messages = []
    date=[]
    time=[]
    senders=[]
    with open(file_path, 'r', encoding = 'utf8') as file:
        for line in file:
            # Use regular expression to extract the message part after ': '
            sender_message_match = re.match(r'.*? - (.*?): (.*)', line)
            if sender_message_match:
                sender, message = sender_message_match.groups()
                senders.append(sender.strip())
                messages.append(message.strip())  
                timestamp = line.split(' - ')[0]
                split_timestamp = timestamp.split(', ')
                date.append(split_timestamp[0].strip())
                time.append(split_timestamp[1].strip())

    df = pd.DataFrame(
        {
            'date': date,
            'time': time,
            'sender': senders,
            'content': messages
        }
    )
    df['date'] = pd.to_datetime(df['date'])
    df['time'] = pd.to_datetime(df['time']).dt.strftime('%H:%M')
    return df
```

```python
person_name = 'Ahmad'
idf = generate_dataframe('Conversations/'+person_name+'.txt')
idf.head()
```

```python
convo_starter = []
convo_starter.append(idf.loc[0]['sender'])
for i in range(1,len(idf)):
    diff = idf.loc[i]['date'] - idf.loc[i-1]['date']
    days = diff.components.days
    if days > 0:
        convo_starter.append(idf.loc[i-1]['sender'])
    else:
        convo_starter.append('in_convo')
idf['convo_starter'] = convo_starter

idf.head()
```

```python
import matplotlib.pyplot as plt
texters = idf['sender'].unique()
convo_starter_df = pd.DataFrame(
    {
        'frequency' : [len(idf[idf['convo_starter'] == texters[0]]), len(idf[idf['convo_starter'] == texters[1]])]
    },
    index = [texters[0], texters[1]]
)

convo_starter_df.plot(kind = 'bar')
plt.show()
```

### Topic Modelling individual

### Text preprocessing

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string

def preprocess_text(text):
    exclude = set(string.punctuation)
    stop_words = set(stopwords.words('english'))
    lemma = WordNetLemmatizer()
    clean = ' '.join([lemma.lemmatize(word.lower()) for word in str(text).split(' ') if word not in stop_words or word not in exclude])
    return clean
```

```python
# getting text data from 'data' dataframe
text_data = data[data['sender'] == 'Ahmad']['content'].tolist()
clean_text = [preprocess_text(text) for text in text_data]
clean_text = [text for text in clean_text if text != '<media omitted>' and '(file attached)' not in text]
clean_text
```

```python
from gensim import corpora 



tokenized_data = [text.split() for text in clean_text]

#create dictionary
id2word = corpora.Dictionary(tokenized_data)


#term document frequency
corpus = [id2word.doc2bow(text) for text in tokenized_data]

#define number of topics
num_topics = 5
```

```python
from gensim.models.ldamodel import LdaModel
from pprint import pprint

model = LdaModel(corpus = corpus, id2word = id2word, num_topics = num_topics, random_state=42, passes=10, alpha='auto', per_word_topics=True)
pprint(model.print_topics())
```

```python
from gensim.models import CoherenceModel
coherence_model_lda = CoherenceModel(model=model, texts=tokenized_data, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```

### Active hours analysis

```python
message_df = pd.DataFrame(
    {
        'time' : data['time'],
        'sender': data['sender'],
        'message': [1 for i in range(len(data['content']))]
    }
)

message_df.head()
```

```python
person_name = 'Elisabeth'
ind_message_df = message_df[message_df['sender'] == person_name].reset_index().drop(columns=['index'])

ind_message_df['time'] = pd.to_datetime(ind_message_df['time'])

ind_message_df.head()
```

```python


def group_time(time):
    intervals = ['00:00', '04:00', '08:00', '12:00', '16:00', '20:00']
    time_intervals = [datetime.strptime(time, '%H:%M') for time in intervals]
    groups = ['00-04', '04-08', '08-12', '12-16', '16-20', '20-00']
    
    if time.hour >= time_intervals[0].hour and time.hour<time_intervals[1].hour:
        return groups[0]
    elif time.hour >= time_intervals[1].hour and time.hour<time_intervals[2].hour:
        return groups[1]
    elif time.hour >= time_intervals[2].hour and time.hour<time_intervals[3].hour:
        return groups[2]
    elif time.hour >= time_intervals[3].hour and time.hour<time_intervals[4].hour:
        return groups[3]
    elif time.hour >= time_intervals[4].hour and time.hour<time_intervals[5].hour:
        return groups[4]
    elif time.hour >= time_intervals[5].hour and time.hour<time_intervals[0].hour:
        return groups[5]
```

```python
ind_message_df['time_period'] = ind_message_df['time'].apply(group_time)
ind_message_df.head()
```

```python
active_hours_df = pd.DataFrame({
    'time_period': ind_message_df['time_period'],
    'message': ind_message_df['message']
})

active_hours = active_hours_df.groupby('time_period')['message'].sum()

active_hours_df = active_hours.to_frame().reset_index()
active_hours_df.head()
```

```python
import matplotlib.pyplot as plt

plt.barh('time_period', 'message', data = active_hours_df)
plt.xlabel('Number of messages')
plt.ylabel('Time periods (hours)')
plt.title(person_name)
plt.show()
```
